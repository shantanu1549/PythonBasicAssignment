A3. In python objects can be classified as Mutable or Immutable based on whether their data can be changed after creation or not.

Mutable Objects:-

1. Modification is possible in place
2. They have methods to alter the object.
3. Examples - List, Dictionary, Set

Used when you need to frequently update data.

Immutable Objects:-

1. Modification is not possible after creation.
2. Methods retunr new objects.
3. Examples - Tuple, String, Integer

Used to maintain data integrity.

```python
'''
Q4. Discuss the different types of operators in Python and provide examples of how they are used?

A4. Python has a variety of operators which perform various operations:-
'''
#1. Arithmetic Operators:-
    #These operators are used to perform mathematical operations such as addition, subtraction, multiplication, and division.

    #Example:-
a = 9
b = 15
print(a + b)  # Addition
print(a - b)  # Subtraction
print(a * b)  # Multiplication
print(a / b)  # Division
print(a % b)  # Modulus (remainder)
print(a // b) # Floor Division (quotient)
print(a ** b) # Exponentiation (power)

#2. Assignment Operators:-
    #These operators are used to assign values to variables.

    #Example:-
a = 15
b = 9
a += b  # a = a + b
a -= b  # a = a - b
a *= b  # a = a * b
a /= b  # a = a / b
a %= b  # a = a % b
a //= b # a = a // b
a **= b # a = a ** b

#3. Comparison Operators:-
    #These operators are used to compare values between variables.

    #Example:-
a = 9
b = 15
print(a == b)  # Equal to (a is equal to b)
print(a != b)  # Not equal to (a is not equal to b)
print(a > b)   # Greater than (a is greater than b)
print(a < b)   # Less than (a is less than b)
print(a >= b)  # Greater than or equal to (a is greater than or equal to b)
print(a <= b)  # Less than or equal to (a is less than or equal to b)

#4. Logical Operators:-
    #These operators are used to combine conditional statements.

    #Example:-
a = 15
b = 9
c = 2
print((a > b) and (c > b))  # True if a > b and c > b
print((a > b) or (c > b))   # True if a > b or c > b
print(not (a > b))          # False if a > b

#5. Membership Operators:-
    #These operators are used to check if a value is present in a sequence.

    #Example:-
a = [15, 4, 9, 17, 11]
print(11 in a)  # True if 11 is in the list a
print(19 not in a)  # True if 19 is not in the list a

#6. Identity Operators:-
    #These operators are used to check if two variables refer to the same object.

    #Example:-
a = 15
b = 9
print(a == b)  # False if a is not equal to b
print(a is b)  # True if a and b refer to the same object
print(a is not b)  # True if a and b do not refer to the same object

#7. Bitwise Operators:-
    #These operators are used to perform bitwise operations on integers.

    #Example:-
a = 15
b = 9
print(a & b)  # Bitwise AND (a and b)
print(a | b)  # Bitwise OR (a or b)
print(a ^ b)  # Bitwise XOR (a xor b)
print(~a)     # Bitwise NOT (~a)
print(a & b)  # Bitwise AND (a and b)
print(a | b)  # Bitwise OR (a or b)
print(a ^ b)  # Bitwise XOR (a xor b)
```

```python
'''
Q5. Explain the concept of type casting in Python with examples?

A5. Type casting is a process of converting a variable from one data type to another.
    Type casting is of 2 types:
'''
    #1. Implicit Type Casting:-
        #Python automatically converts one data type to another without any user involvement.

        #Example:-

a = 15
b = 9.4
c = a + b
print(c)
print(type(c))

#Output:-
#24.4
#<class 'float'>

    #2. Explicit Type Casting:-
        #Python performs it by using predefined functions to convert from one data type to another.

        #Example:-

a = 15
b = 9.4
c = a + int(b)
print(c)
print(type(c))

#Output:-
#24
#<class 'int'>

#Example:-

a = 15
b = 9.4
c = float(a) + b
print(c)
print(type(c))

#Output:-
#24.4
#<class 'float'>

#Example:-

a = 15
b = 9.4
c = int(a) + int(b)
print(c)
print(type(c))

#Output:-
#24
#<class 'int'>

#Example:-

a = 15
b = 9.4
c = str(a) + str(b)
print(c)
print(type(c))

#Output:-
#159.4
#<class 'str'>

#Example:-

l = [4, 9, 15]
t = (11, 10, 5)
c = l + list(t)
print(c)
print(type(c))

#Output:-
#[4, 9, 15, 11, 10, 5]
#<class 'list'>
```
